# Changing the number of minor breaks

```python
import pandas as pd
import numpy as np
from plotnine import *
from plotnine.data import mtcars
from mizani.breaks import minor_breaks
from mizani.transforms import identity_trans

%matplotlib inline
```

Set the default theme to one that makes the minor breaks for the y-axis stand out.

```python
original_theme = theme_set(
    theme_gray()
    + theme(panel_grid_minor_y=element_line(color='brown', linetype='dashed', alpha=.5))
)
```

By default, there is one minor break between any two major breaks.

```python
(ggplot(mtcars)
 + geom_point(aes('wt', 'mpg'))
)
```

The minor breaks are part of the transformation for any given aesthetic. All of which is managed by the
underlying scales framework `mizani`. Scales allow you to override them with specific values or a
a function to calculate minor breaks. Passing a function to the scales, the minor breaks are calculated
globally.

```python
(ggplot(mtcars)
 + geom_point(aes('wt', 'mpg'))
 + scale_y_continuous(minor_breaks=minor_breaks(4))
)
```

In this case, that is not what we want.

**Solution**

What we need to do is pass the function to the transform. 

```python
(ggplot(mtcars)
 + geom_point(aes('wt', 'mpg'))
 + scale_y_continuous(trans=identity_trans(minor_breaks=minor_breaks(4)))
)
```
